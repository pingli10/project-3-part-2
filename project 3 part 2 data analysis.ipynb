```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_20 = pd.read_csv('Data/final_tmdb_data_2000.csv.gz')
df_21 = pd.read_csv('Data/final_tmdb_data_2001.csv.gz')
```

```python
df = pd.concat([df_20,df_21])
```

```python
# see revenue of movies by each certification
sns.barplot(data=df,x='certification',y='revenue')
```

```python
# check boxplot by x = Drug y = age
plt.figure(figsize=(10,7))
sns.boxplot(x='certification',y='revenue',data=df);
```
